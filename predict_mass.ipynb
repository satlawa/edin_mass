```python
###########################
###   preprocess data   ###
###########################

data_thin = shuffle(data_thin, random_state=399)

# filter for experiments
train = data_thin[:int(data_thin.shape[0]*0.8)]
test = data_thin[int(data_thin.shape[0]*0.8):]

X_train = train.drop(['cut_lh_ha', 'cut_nh_ha', 'cut_sum_ha'], axis=1).values
y_train = train['cut_sum_ha'].values
X_test = test.drop(['cut_lh_ha', 'cut_nh_ha', 'cut_sum_ha'], axis=1).values
y_test = test['cut_sum_ha'].values

print("Train: number of instances: {}, number of attributes: {}".format(train.shape[0], train.shape[1]))
print("Test: number of instances: {}, number of attributes: {}".format(test.shape[0], test.shape[1]))
```

```python
############################
###   standardize data   ###
############################

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
##########################
##########################
###                    ###
###   Classification   ###
###                    ###
##########################
##########################
```

```python
###########################################
###   SVC - Support Vector Classifier   ###
###########################################
```

```python
clf = SVR(kernel='rbf', C=100, gamma='auto')
scores = cross_val_score(clf, X_train, y_train, cv=10, n_jobs=6)
print(np.mean(scores), scores)
```

```python
param_grid = [
    {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000], 'gamma': ['scale', 'auto'], 'kernel': ['rbf']},
    {'C': [1, 10, 100, 1000], 'gamma': ['scale', 'auto'], 'degree': [2, 3, 4], 'kernel': ['poly']},
    {'C': [1, 10, 100, 1000], 'gamma': ['scale', 'auto'], 'kernel': ['sigmoid']},
 ]
```

```python
clf = GridSearchCV(SVR(), param_grid, cv=10, n_jobs=6)
clf.fit(X_train, y_train)
```

```python
print("------------------------------------------------------------------------------------------------------")
print()
print("Best parameters set found on development set: ", clf.best_params_)
print()
print('Best mean cross-validated score:', clf.best_score_)
print()
print("------------------------------------------------------------------------------------------------------")
print("Grid scores for SVM:")
print()
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
          % (mean, std * 2, params))
```
